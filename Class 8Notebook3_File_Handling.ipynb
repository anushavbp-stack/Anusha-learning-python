## 1. Opening Files — The `open()` Function

```python
file = open("filename.txt", mode)
```

| Mode | Meaning | Creates file? | Overwrites? |
|------|---------|--------------|-------------|
| `'r'` | Read (default) | ❌ | ❌ |
| `'w'` | Write | ✅ | ✅ |
| `'a'` | Append | ✅ | ❌ |
| `'x'` | Create (fails if exists) | ✅ | ❌ |
| `'r+'` | Read + Write | ❌ | ❌ |

> ⚠️ Always close files after use. The `with` statement does this automatically.

## 2. Reading Files Line by Line

For large files, reading line by line is memory-efficient.

````python
# Method 1: readlines() — returns a list of lines
with open("anushap.txt", "r") as f:
    line1 = f.readline()
    line2 = f.readline()
    line3 = f.readline()
    

print(line1, line2, line3)

for i, lines in enumerate([line1, line2, line3], 1):
     print(f"Line {i}: {lines.strip()}") #we use .strip() to avoid the extra blank lines.
     #print(f"Total lines: {len(lines)}")
````

````python
# Method 2: iterate directly (most memory-efficient)
with open("anushap.txt", "r") as f:
    for line in f:
        print(f"{line}")# in this case we do not need  "f" while printing.
        
````

## 3. Appending to Files

Use `'a'` mode to add content without overwriting existing data.

````python
# Append new lines
with open("anusha.txt", "a") as f:
    f.write("new line is added.\n")
    f.write("using 'a' mode for appending\n")

# Verify
with open("anusha.txt", "r") as f:
    print(f.read())
````

## 4. Writing Multiple Lines with `writelines()`

````python
students = ["Alice\n", "Bob\n", "Charlie\n", "Diana\n"]

with open("mantha.txt", "w") as f:
    f.writelines(students)

# Read back
with open("mantha.txt", "r") as f:
    for line in f:
        print(line.strip())
````

## 6. Exception Handling for Files

````python
# TRY AND ERROR METHOD IN FILE HANDLING 

files = {'file_1' : "mantha.txt", 'file_2' : "mantha1.txt"}

try:
    files['file_3']  # this will raise a KeyError because 'file_3' is not defined in the dictionary.
except KeyError as e:   #KeyError is raised when we try to access a key that does not exist in the dictionary.
                        # e is the variable that holds the error message, which we can print to understand what went wrong.
    print(f"There is a key error in the above code - {e}")
except Exception as e:       # exception means any error that is not a KeyError will be caught here.
    print(f"There is some error in the above code - {e}")

````

````python
# Handle missing files gracefully
def safe_read(anuu):
    try:
        with open(anuu, "r") as f:
            return f.read()
    except FileNotFoundError:
        return f"❌ Error: '{anuu}' not found."
    except PermissionError:
        return f"❌ Error: No permission to read '{anuu}'."
    except Exception as e:
        return f"❌ Unexpected error: {e}"

print(safe_read("anuu.txt"))
print(safe_read("ghost_file.txt"))
````

````python
# Exercise 1: Word Counter
# Write a function count_words(filepath) that:
# - Reads a text file
# - Returns a dict with each unique word as key and its count as value
# - Is case-insensitive ("The" and "the" count as the same word)
# - Ignores punctuation

import string

def count_words(filepath):
    word_count = {}
    with open(filepath, "r") as f:
        for line in f:
            # Convert to lowercase and remove punctuation
            line = line.lower().translate(str.maketrans('', '', string.punctuation))
            words = line.split()
            for word in words:
                word_count[word] = word_count.get(word, 0) + 1
    return word_count

# Create a test file first
with open("test_words.txt", "w") as f:
    f.write("the cat sat on the mat\nthe cat is a fat cat\n")

result = count_words("test_words.txt")
for word, count in sorted(result.items()):
    print(f"  {word}: {count}")
````

````python
import string

def count_words(myfilepath):
    word_count = {}

    with open(myfilepath, "r") as f:
        content=f.read()
    content = content.lower()  # Convert to lowercase for case-insensitivity
     # Remove punctuation
    for char in string.punctuation:
        content = content.replace(char, "")

    # Split text into words
    words = content.split()

    # Count words
    for word in words:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1

    return word_count


# Create a sample file
with open("myfile.txt", "w") as f:
    f.write("compute is a big innovation in the world. compute is used everywhere.")

# Call function
result = count_words("myfile.txt")

print(result)
````
